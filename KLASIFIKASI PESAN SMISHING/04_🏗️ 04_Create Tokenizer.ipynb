## âœ¨ 3.1 - Create Char Level Corpus for Tokenizing and Padding

```python
X = [str(text) for text in X]

tokenizer_char_level = Tokenizer(num_words=None, oov_token='<OOV>', char_level=True)
tokenizer_char_level.fit_on_texts(X)
# X_sequences_char_level = tokenizer_char_level.texts_to_sequences(X)
# X_padded_char_level = pad_sequences(X_sequences_char_level, padding='post', truncating='post', maxlen=2000)

# # Build vocabulary
# vocab_size_char_level = len(tokenizer_char_level.word_index) + 1

# X_padded_char_level
```

```python
tokenizer_char_json = tokenizer_char_level.to_json()

# Menyimpan representasi JSON ke dalam sebuah file
with open('/content/drive/MyDrive/TUGAS AKHIR/Notebook/PESAN TEKS/Tokenizer/tokenizer_char_level.json', 'w', encoding='utf-8') as f:
    f.write(json.dumps(tokenizer_char_json))
```
